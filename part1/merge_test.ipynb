```python
import pandas as pd
df1=pd.DataFrame({'name':['kate','herz','catherine','sally','yyh'],
'age':[25,28,39,35,100]})

df4=pd.DataFrame({'name':['kate','herz','catherine','sally','sally','yyh','yyh'],
'age':[25,28,39,35,36,100,80],'time':[1,2,3,4,5,6,7]})

df2=pd.DataFrame({'name':['kate','herz','sally'],
'score':[70,60,90]})

df3=pd.DataFrame({'name':['kate','herz','sally','cristin'],
'score':[70,60,90,30]})

pd.merge(df1,df2)
```
